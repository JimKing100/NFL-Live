```
# Imports
import pandas as pd
```

```
# Load the data
players_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/NFL-Live/master/data-science/data/raw/players_full.csv')
bye_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/NFL-Live/master/data-science/data/raw/bye.csv')
injury_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/NFL-Live/master/data-science/data/raw/injury.csv')
```

```
# The dataframe of all 2019 players
players_df.head()
```

```
# The dataframe of bye weeks for 2019 teams
bye_df.head()
```

```
# The dataframe of injuries for 2019 players
injury_df.head()
```

```
# The main code for iterating through the player list and adding the bye week and injury week
players_list = players_df['player'].tolist()
bye_week = []
injury_week = []

for player in players_list:

    team = players_df['cteam'].loc[(players_df['player']==player)].iloc[0]

    bye_week.append(bye_df['bye-week'].loc[(bye_df['player']==team)].iloc[0])
    injury_weeks = injury_df['week'].loc[(injury_df['player']==player)]
    
    # If no injuries set injury week to non-existing week 18
    if (len(injury_weeks)==0):
        injury_week.append(18)
    else:
        injury_week.append(min(injury_weeks))

players_df['bye_week'] = bye_week
players_df['injury_week'] = injury_week 
```

```
# The revised players_df dataframe
players_df
```

```
# Save the results to .csv file
players_df.to_csv('/content/players_full1.csv', index=False)
```
